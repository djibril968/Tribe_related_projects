# Data Cleaning

```python
#renaming and dropping unwanted column
tribe_data = data
tribe_data.rename(columns= {"Status" : "Gender"}, inplace= True)
tribe_data.drop(columns= ["name", "Why?"], inplace= True)

tribe_data
```

```python
#Now we expand our date columns 
tribe_data['Year'] = pd.to_datetime(tribe_data['Event Date']).dt.year
tribe_data['Month'] = pd.to_datetime(tribe_data['Event Date']).dt.month
tribe_data['Month_name'] = pd.to_datetime(tribe_data['Event Date']).dt.month_name()
```
