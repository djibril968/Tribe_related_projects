```python
#preferred event day by gender
event_data_exploded = event_data.assign(
    pref_days=event_data['What day of the week are most convenient for you to attend events?']
    .str.split(',')
).explode('pref_days')

# here we remove any leading or trailing whitespace from the new column
event_data_exploded['pref_days'] = event_data_exploded['pref_days'].str.strip()
event_data_exploded

#to get the preferred event days of the week grouped by gender
pref_event_day_by_gender = event_data_exploded.groupby('Gender')['pref_days'].value_counts().reset_index(name= 'counts')
pref_event_day_by_gender 
```

```python
day_pref.rename(columns={'What day of the week are most convenient for you to attend events?': 'week_day'}, inplace= True)
```

```python
def replace_dayname(val):
    if val == 'Sundays':
        return 'Sun'
    if val == 'Mondays':
        return 'Mon'
    if val == 'Tuesdays':
        return 'Tue'
    if val == 'Wednesdays':
        return 'Wed'
    if val == 'Thursdays':
        return 'Thu'
    if val == 'Fridays':
        return 'Fri'
    if val == 'Saturdays':
        return 'Sat'
    else:
        return val

day_pref['week_day'] = day_pref['week_day'].apply(replace_dayname)
day_pref
```

```python
plt.bar(day_pref['week_day'], day_pref['counts'])
plt.title('Most preferred day of event')
plt.xlabel('weekday')
plt.show()
```
